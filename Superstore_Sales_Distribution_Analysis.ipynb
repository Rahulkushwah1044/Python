```python
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(), inplace = True)
```

```python
plt.figure(figsize = (6,3))
sns.barplot(x='Product Category',y = 'Profit', data = df, estimator = 'sum')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['year'] = df['Order Date'].dt.year
df['Order Year'] = df['Order Date'].dt.year
```

```python
df['Order Year'].value_counts()
```

- Measure the distribution with countplot of Order Year.

```python
plt.figure(figsize = (5,4))
S = sns.countplot(x = 'Order Year', data = df)
S.bar_label(S.containers[0])
plt.title('CountPlot of Order year')
plt.show()
```

```python
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values
```

- Measure the percentage of ship mode.

```python
plt.figure(figsize=(6,5))
#s = df.groupby('Ship Mode').agg({'Ship Mode':'count'})
plt.pie(y,autopct = '%1.2f%%', labels = x,explode=[0.01,0.01,0.01],startangle=60)
plt.legend(loc=2)
plt.title('Ship Mode')
plt.show()
```

**Order Priority**

```python
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority'] = df['Order Priority'].replace('Critical','Critical')
```

- Measure the order priority with help of countplot.

```python
plt.figure(figsize = (6,5))
S = sns.countplot(x = 'Order Priority', data = df, hue = True )
S.bar_label(S.containers[0])
plt.title('Count of Order Priority')
plt.show()
```

- Making the pie chart of percentage of order priority.

```python
plt.figure(figsize=(8,5))
p = df.groupby('Order Priority').agg({'Order Priority':'count'})
plt.pie(p['Order Priority'], labels = p.index, autopct = '%1.2f%%',explode=[0.01,0.01,0.01,0.01,0.01,0.01])
plt.title('Percentage of Order Priority', fontsize = 15)
plt.show()
```
